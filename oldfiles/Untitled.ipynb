```python
import pandas as pd
import numpy as np
```

```python
imdb = pd.read_csv('resources/IMDb_movies.csv')
imdb.head(5)
```

```python
imdb_df = imdb.drop(columns=['date_published', 'genre', 'description', 'original_title', 'duration', 'language', 'actors', 'avg_vote', 'votes', 'budget', 'usa_gross_income', 'worlwide_gross_income', 'metascore', 'reviews_from_users', 'reviews_from_critics'])
imdb_df.head(5)
```

```python
# Conver tuple in multiple value column to list
imdb_df2 = imdb_df.assign(director=imdb_df.director.str.split(','))
imdb_df2
```

```python
# Using Pandas explode to separate director names into another column
imdb_df3 = imdb_df2.explode('director')
imdb_df3
```
